```python
import tensorflow as tf
from tensorflow import keras
import pandas as pd
import numpy as np
```

```python
train_dataset = pd.read_csv("train.csv")
```

```python
np.random.shuffle(train_dataset.values)
```

```python
from tensorflow.keras.layers import Dense
```

```python
model = keras.Sequential([
    Dense(8, input_shape = (2,), activation="relu"),
    Dense(2, activation="sigmoid")
])
```

```python
from tensorflow.keras.losses import SparseCategoricalCrossentropy
```

```python
model.compile(optimizer="adam", loss=keras.losses.SparseCategoricalCrossentropy(from_logits=True), metrics = ["accuracy"])
```

```python
x = np.column_stack((train_dataset.x.values, train_dataset.y.values))
```

```python
model.fit(x,train_dataset.color.values ,batch_size=4, epochs=5)
```

```python
test_df = pd.read_csv("test.csv")
```

```python
test_x = np.column_stack((test_df.x.values, test_df.y.values))
```

```python
loss, accuracy = model.evaluate(test_x, test_df.color.values)
```

```python
print(f"loss:{loss:.6f}, accuracy:{accuracy:.2f}")
```
