```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
from torch.utils.data import DataLoader, TensorDataset

from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
X, y = load_breast_cancer(return_X_y=True)
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2)
scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
X_test_scaled
```

```python
y_train
```

```python
X_train_scaled_tensor = torch.from_numpy(X_train_scaled).float()
X_test_scaled_tensor = torch.from_numpy(X_test_scaled).float()
y_train_tensor = torch.from_numpy(y_train).float().unsqueeze(1)
y_test_tensor = torch.from_numpy(y_test).float().unsqueeze(1)
```

```python
train_dataset = TensorDataset(X_train_scaled_tensor, y_train_tensor)
```

```python
X_train_scaled_tensor.shape
```

```python
train_loader = DataLoader(train_dataset, batch_size = 32, shuffle = True)
```

```python
class BCNet(nn.Module):
  def __init__(self):
    super(BCNet, self).__init__()
    self.fc1 = nn.Linear(30, 128)
    self.fc2 = nn.Linear(128, 64)
    self.fc3 = nn.Linear(64,32)
    self.fc4 = nn.Linear(32,1)
    self.relu = nn.ReLU()
    self.sigmoid = nn.Sigmoid()

  def forward(self, x):
    x = self.fc1(x)
    x = self.relu(x)
    x = self.fc2(x)
    x = self.relu(x)
    x = self.fc3(x)
    x = self.relu(x)
    x = self.fc4(x)
    x = self.sigmoid(x)
    return x
```

```python
model = BCNet()
```

```python
criterio = nn.BCELoss()
optimizer = optim.Adam(model.parameters(),lr = 0.0001)
```

```python
epochs = 20

for epoch in range(epochs):
  model.train()
  running_loss = 0.0

  for x_batch, y_batch in train_loader:
    optimizer.zero_grad()
    preds = model(x_batch)
    loss = criterio(preds, y_batch)

    loss.backward()
    optimizer.step()

    running_loss += loss.item()
  print(f"Epoch {epoch+1}: Loss was {running_loss/ len(train_loader)}")
```

```python
with torch.no_grad():
  model.eval()

  preds = model(X_test_scaled_tensor)
  loss = criterio(preds, y_test_tensor).item()
  accuracy = ((preds >= 0.5) == y_test_tensor).float().mean().item()
```
